# Matplotlib tutorial

- Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python.
- It is a Python library used for plotting graphs with the help of other libraries like Numpy and Pandas. It is a powerful tool for visualizing data in Python. 
- It is used for creating statical interferences and plotting 2D graphs of arrays. It was first introduced by John D. Hunter in 2002. 
- It uses Pyplot for providing MATLAB like interface free and open-source. It is capable of dealing with various operating systems and their graphical backends.

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [1,2,3,4,5,6,7]
y = [50,51,52,48,47,49,46] 
```

```python
# Line graph
plt.plot(x,y)  # plot x and y using default line style and color
```

```python
plt.figure(figsize = (10,6))     
plt.plot(x,y, 'm--.',markersize = 20, alpha = 0.7)  # A format string consists of a part for color, marker and line.
plt.xlabel('Days')           # x-axis label
plt.ylabel('Teperature(C)')  # y-axis label
plt.title('Weather')         # title of the figure
plt.show()
```

## Axes labels, legend, Grid

```python
days= [1,2,3,4,5,6,7]
max_t = [50,51,52,48,47,49,46]
min_t = [43,42,40,44,33,35,37]
avg_t = [45,48,48,46,40,42,41]
```

```python
plt.plot(x, max_t,'r-*',markersize = 10, label = 'max')
plt.plot(x,min_t, 'm--+',markersize = 15, label = 'min')
plt.plot(x,avg_t, 'b:o', label = 'avg')
plt.xlabel('Days')
plt.ylabel('Temperature(c)')
plt.title('Weather')

plt.legend(loc = 'best', shadow = True, fontsize = 'medium')

plt.grid()
```

## Bar Charts

```python
import numpy as np
```

```python
company = ['GOOGL','AMZN','MSFT','FB']
revenue = [90, 136, 89, 27]
profit = [40,2,34,12]
```

```python
xpos = np.arange(len(company))
xpos
```

```python
plt.bar(xpos-0.2, revenue, width = 0.4, label = 'revenue')
plt.bar(xpos+0.2, profit, width =0.4,label = 'profit' )
plt.xticks(xpos, company)
plt.xlabel('Companies')
plt.ylabel('Revenue (billion)')
plt.title('US Tech Stocks')
plt.legend()
plt.show()
```

If we want to horizonal bar charts instead of vertical, we can use **barh** in place of **bar**.

## Histogram

**Blood Sugar level**
- 80 - 100  : Normal
- 100 - 125 : Pre-diabetic
- 125 above : Diabetic

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
```

```python
plt.hist(blood_sugar, bins =[80,100,125,150], rwidth = 0.95)
```

```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]
```

```python
plt.hist([blood_sugar_men, blood_sugar_women], bins =[80,100,125,150], label = ['men','women'])
plt.xlabel('Sugar range')
plt.ylabel('No. of patients')
plt.title('Blood Sugar Analysis ')
plt.legend()
plt.show()
```

## Pie Chart

```python
exp_values = [1400, 600, 300, 410, 250]
exp_labels = ['Home rent', 'Food', 'Phone/internet bill','Car','Other Utilities']
```

```python
plt.pie(exp_values, labels = exp_labels, radius = 1.6, autopct ='%0.2f%%', explode =[0,0,0,0.1,0.3], shadow = True)
plt.show()
```

## Save chart to a image file

```python
plt.pie(exp_values, labels = exp_labels, radius = 1.6, autopct ='%0.2f%%', explode =[0,0,0,0.1,0.3], shadow = True)
plt.savefig('piechart.png', bbox_inches = 'tight', pad_inches = 2)
```
